# Support Vector Machine
: Multi-propose machine learning model which can be used for linear or non-linear classification, regression and outlier detection. Support vector machine is suitable for <b>complicated classification problem and dim-size dataset.</b>

# Linear SVM Classification
1. There are many support vectors that can classify our sample. -> classifier with <b>large margin is better</b> than others with small margin.
2. SVM is sensitive to the <b>scale of data.</b>

### Small Margin vs Large Margin

```python
#from https://github.com/rickiepark/handson-ml2
from sklearn.svm import SVC
from sklearn import datasets

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # length, width
y = iris["target"]

setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]

# SVM 분류 모델
svm_clf = SVC(kernel="linear", C=float("inf"))
svm_clf.fit(X, y)
```

```python
#from https://github.com/rickiepark/handson-ml2
import matplotlib.pyplot as plt
import numpy as np

x0 = np.linspace(0, 5.5, 200)
pred_1 = 5*x0 - 20
pred_2 = x0 - 1.8
pred_3 = 0.1 * x0 + 0.5

def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]

    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]

    margin = 1/w[1]
    upper = decision_boundary + margin
    lower = decision_boundary - margin

    svs = svm_clf.support_vectors_
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, upper, "k--", linewidth=2)
    plt.plot(x0, lower, "k--", linewidth=2)

fig, axes = plt.subplots(ncols=2, figsize=(10,3), sharey=True)

plt.sca(axes[0])
plt.title('Small Margin Classification')
plt.plot(x0, pred_1, "g--", linewidth=2)
plt.plot(x0, pred_2, "m-", linewidth=2)
plt.plot(x0, pred_3, "r-", linewidth=2)
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris versicolor")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris setosa")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 5.5, 0, 2])

plt.sca(axes[1])
plt.title('Large Margin Classification')
plot_svc_decision_boundary(svm_clf, 0, 5.5)
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo")
plt.xlabel("Petal length", fontsize=14)
plt.axis([0, 5.5, 0, 2])

plt.show()
```

### Sensitivity to Scale

```python
#from https://github.com/rickiepark/handson-ml2
Xs = np.array([[1, 50], [5, 20], [3, 80], [5, 60]]).astype(np.float64)
ys = np.array([0, 0, 1, 1])
svm_clf = SVC(kernel="linear", C=100)
svm_clf.fit(Xs, ys)

plt.figure(figsize=(9,3))
plt.subplot(121)
plt.plot(Xs[:, 0][ys==1], Xs[:, 1][ys==1], "bo")
plt.plot(Xs[:, 0][ys==0], Xs[:, 1][ys==0], "ms")
plot_svc_decision_boundary(svm_clf, 0, 6)
plt.xlabel("$x_0$", fontsize=20)
plt.ylabel("$x_1$    ", fontsize=20, rotation=0)
plt.title("Unscaled", fontsize=16)
plt.axis([0, 6, 0, 90])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(Xs)
svm_clf.fit(X_scaled, ys)

plt.subplot(122)
plt.plot(X_scaled[:, 0][ys==1], X_scaled[:, 1][ys==1], "bo")
plt.plot(X_scaled[:, 0][ys==0], X_scaled[:, 1][ys==0], "ms")
plot_svc_decision_boundary(svm_clf, -2, 2)
plt.xlabel("$x'_0$", fontsize=20)
plt.ylabel("$x'_1$  ", fontsize=20, rotation=0)
plt.title("Scaled", fontsize=16)
plt.axis([-2, 2, -2, 2])
```

## Large Margin vs Margin Violation

### Hard Margin
* <b>Hard margin classifier is sensitive to outliers</b>
* <b>Hard margin classifier can work when data can be separate in two groups</b>


```python
#from https://github.com/rickiepark/handson-ml2
X_outliers = np.array([[3.4, 1.3], [3.2, 0.8]])
y_outliers = np.array([0, 0])
Xo1 = np.concatenate([X, X_outliers[:1]], axis=0)
yo1 = np.concatenate([y, y_outliers[:1]], axis=0)
Xo2 = np.concatenate([X, X_outliers[1:]], axis=0)
yo2 = np.concatenate([y, y_outliers[1:]], axis=0)

svm_clf2 = SVC(kernel="linear", C=10**9)
svm_clf2.fit(Xo2, yo2)
fig, axes = plt.subplots(ncols=2, figsize=(10,2.7), sharey=True)

plt.sca(axes[0])
plt.plot(Xo1[:, 0][yo1==1], Xo1[:, 1][yo1==1], "bs")
plt.plot(Xo1[:, 0][yo1==0], Xo1[:, 1][yo1==0], "yo")
plt.text(0.3, 1.0, "Impossible!", fontsize=24, color="red")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.annotate("Outlier",
             xy=(X_outliers[0][0], X_outliers[0][1]),
             xytext=(2.5, 1.7),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=16,
            )
plt.axis([0, 5.5, 0, 2])

plt.sca(axes[1])
plt.plot(Xo2[:, 0][yo2==1], Xo2[:, 1][yo2==1], "bs")
plt.plot(Xo2[:, 0][yo2==0], Xo2[:, 1][yo2==0], "yo")
plot_svc_decision_boundary(svm_clf2, 0, 5.5)
plt.xlabel("Petal length", fontsize=14)
plt.annotate("Outlier",
             xy=(X_outliers[1][0], X_outliers[1][1]),
             xytext=(3.2, 0.08),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=16,
            )
plt.axis([0, 5.5, 0, 2])

plt.show()
```

### Soft Margin
By controlling hyperparameter C, we can regulate the SVM model. The less C value is, the more model is regulated. In hard margin classifier, C is infinite.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn import datasets
from sklearn.pipeline import Pipeline

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # length, width
y = iris["target"] == 2      # Is iris-virginica?

svc_pipeline = Pipeline([
    ('std_scaler', StandardScaler()),
    ('lin_svc', LinearSVC(C = 1, loss = 'hinge'))
])

svc_pipeline.fit(X, y)
svc_pipeline.predict([[5.5, 1.7]])
```

```python
std_scaler = StandardScaler()
X_std = std_scaler.fit_transform(X)

lin_svc_1 = LinearSVC(C = 1, loss = 'hinge', random_state = 42)
lin_svc_100 = LinearSVC(C = 100, loss = 'hinge', random_state = 42)
lin_svc_1.fit(X_std, y)
lin_svc_100.fit(X_std, y)

svc_pipe1 = Pipeline([
    ('std', std_scaler),
    ('svc1', lin_svc_1)
])
svc_pipe2 = Pipeline([
    ('std', std_scaler),
    ('svc100', lin_svc_100)
])


plot_svc_decision_boundary(svc_pipe1, 4, 6)
plot_svc_decision_boundary(svc_pipe2, 4, 6)

plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^", label="Iris virginica")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs", label="Not virginica")
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10,2.7), sharey=True)

plt.sca(axes[0])
plot_svc_decision_boundary(svm_clf1, 4, 5.9)
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.title("$C = {}$".format(svm_clf1.C), fontsize=16)
plt.axis([4, 5.9, 0.8, 2.8])

plt.sca(axes[1])
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
plot_svc_decision_boundary(svm_clf2, 4, 5.99)
plt.xlabel("Petal length", fontsize=14)
plt.title("$C = {}$".format(svm_clf2.C), fontsize=16)
plt.axis([4, 5.9, 0.8, 2.8])
```

# Non-linear SVM Classification
Even though dataset cannot be seperate linearly, we can classify the data by adding higher degree feature.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.datasets import make_moons

X_raw, y_raw = make_moons(n_samples = 100, noise = 0.15, random_state=42)
plt.scatter(X_raw[:,0][y_raw == 0], X_raw[:,1][y_raw == 0], label = 'y is 0')
plt.scatter(X_raw[:,0][y_raw == 1], X_raw[:,1][y_raw == 1], label = 'y is 1')
plt.legend()
```

```python
poly_svm_pipe = Pipeline([
    ('poly', PolynomialFeatures(degree = 3, include_bias = False)),
    ('std', StandardScaler()),
    ('svc', LinearSVC(C = 10, loss = 'hinge', max_iter = 5000))
])
poly_svm_pipe.fit(X_raw, y_raw)
poly_svm_pipe.predict([[2,2]])
```

```python
def plot_contour(model, min_x1, max_x1, min_x2, max_x2, X_raw, y_raw, m = 100):
    x1s = np.linspace(min_x1, max_x1, m)
    x2s = np.linspace(min_x2, max_x2, m)
    x1, x2 = np.meshgrid(x1s, x2s)
    X = np.c_[x1.ravel(), x2.ravel()]
    y_pred = model.predict(X).reshape(m, m)
    y_decision = model.decision_function(X).reshape(m, m)
    plt.contourf(x1, x2, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x1, x2, y_decision, cmap=plt.cm.brg, alpha=0.1)
    plt.scatter(X_raw[:,0][y_raw == 0], X_raw[:,1][y_raw == 0], label = 'y is 0')
    plt.scatter(X_raw[:,0][y_raw == 1], X_raw[:,1][y_raw == 1], label = 'y is 1')
    plt.legend()
    
plot_contour(poly_svm_pipe, -1.5, 2.5, -1, 1.5, X_raw, y_raw, m = 100)
```

## Polynomial Kernel
Using kernel trick, we can have an effect of adding higher degree attributes without adding more attributes actually.

```python
from sklearn.svm import SVC

fog = plt.figure(figsize = (16, 6))
degrees = [3, 10]
rs = [1, 100]
plot_idx = 1
for deg, r in zip(degrees, rs):
    kernel_svm_pipe = Pipeline([
        ('std', StandardScaler()),
        ('svc', SVC(C = 5, kernel = 'poly', degree = deg, coef0 = r)) # coef0 is r
    ])
    kernel_svm_pipe.fit(X_raw, y_raw)
    plt.subplot(1, 2, plot_idx)
    plt.title('d= {0}, r = {1}, C = 5'.format(deg, r))
    plot_contour(kernel_svm_pipe, -1.5, 2.5, -1, 1.5, X_raw, y_raw, m = 100)
    plot_idx += 1
```

## Similarity Property
By adding a new property calculated using similarity function, we can handle non-linear data set.
* similarity function measures how each sample is close to the specific landmark

### Gaussian RBF
$\phi_{\gamma}(x,l) = exp(-\gamma||x-l||^2)$ where $l$ is landmark
* Lankmark choice: Simple way is to set landmarks in each sample. Then too much attributes will be generated. If # of samples is $m$, $m$ new attributes will be added.
* As $\gamma$ increases, gaussian rbf kernel gets narrower. This change results in more dense distribution of similarity.

```python
def gauss_rbd(x, l, gamma):
    l2_dist = np.linalg.norm(x - l, axis = 1)
    return np.exp(-gamma * np.square(l2_dist).reshape(-1, 1))
```

```python
LANDMARK_X2 = np.array(-2).reshape(-1,1)
LANDMARK_X3 = np.array(1).reshape(-1,1)

X1D = np.linspace(-4, 4, 9).reshape(-1, 1)
y = np.array([0, 0, 1, 1, 1, 1, 1, 0, 0])
X_new = np.linspace(-4.5, 4.5, 100).reshape(-1, 1)

fig = plt.figure(figsize=(14, 9)) 
GAMMAS = [0.01, 0.1, 0.3, 0.9]
plot_idx = 1
for GAMMA in GAMMAS:
    rbf_x2_plot = gauss_rbd(X_new, LANDMARK_X2, GAMMA)
    rbf_x3_plot = gauss_rbd(X_new, LANDMARK_X3, GAMMA)
    plt.subplot(2, 2, plot_idx)
    plt.title('$\gamma$: {0}'.format(GAMMA))
    plt.xlabel('X1')
    plt.ylabel('Similarity')
    plt.grid(True, which='both')
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.plot(X_new, rbf_x2_plot, 'g--', label = 'X2')
    plt.plot(X_new, rbf_x3_plot, 'y--', label = 'X3')
    plt.scatter(X1D[y == 0], np.zeros(len(X1D[y == 0])), color = 'b')
    plt.scatter(X1D[y == 1], np.zeros(len(X1D[y == 1])), color = 'r')
    plt.legend()
    plot_idx += 1
```

```python
fig = plt.figure(figsize=(14, 9)) 

GAMMAS = [0.01, 0.1, 0.3, 0.9]
plot_idx = 1
for GAMMA in GAMMAS:
    rbf_x2 = gauss_rbd(X1D, LANDMARK_X2, GAMMA)
    rbf_x3 = gauss_rbd(X1D, LANDMARK_X3, GAMMA)

    plt.subplot(2, 2, plot_idx)
    plt.title('$\gamma$: {0}'.format(GAMMA))
    plt.xlabel('X2 Similarity')
    plt.ylabel('X3 Similarity')
    plt.grid(True, which='both')
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    if GAMMA == 0.3:
        plt.plot([0, 1], [0.6, -0.1], "k--", linewidth=3)
    plt.scatter(rbf_x2[y == 0], rbf_x3[y == 0], color = 'b')
    plt.scatter(rbf_x2[y == 1], rbf_x3[y == 1], color = 'r')
    plot_idx += 1
```

## Mid-tern Summary
* Kernel funcion is very critical in SVC. It can help model to 
* Small $\gamma$, small C reglates model more.
* There are some kernels that specializes in some data(string kernel, Levenshtein distance function ...). However rbf is most frequently used.
### Which kernel should be used?
1. Try linearSVC first (fastest)
2. Then use SVC(kernel = 'linear')
3. If the train set is not too big, use rbf kernel.

<b>Tip: difference betwwen LinearSVC and SVC(kernel = linear)</b>
from(https://stackoverflow.com/questions/45384185/what-is-the-difference-between-linearsvc-and-svckernel-linear)
* By default scaling, LinearSVC minimizes the squared hinge loss while SVC minimizes the regular hinge loss. It is possible to manually define a 'hinge' string for loss parameter in LinearSVC.
* LinearSVC uses the One-vs-All (also known as One-vs-Rest) multiclass reduction while SVC uses the One-vs-One multiclass reduction. It is also noted here. Also, for multi-class classification problem SVC fits N * (N - 1) / 2 models where N is the amount of classes. LinearSVC, by contrast, simply fits N models. If the classification problem is binary, then only one model is fit in both scenarios. multi_class and decision_function_shape parameters have nothing in common. The second one is an aggregator that transforms the results of the decision function in a convenient shape of (n_features, n_samples). multi_class is an algorithmic approach to establish a solution.
* The underlying estimators for LinearSVC are <b>liblinear, that do in fact penalize the intercept.</b> SVC uses <b>libsvm estimators that do not.</b> liblinear estimators are optimized for a linear (special) case and thus converge faster on big amounts of data than libsvm. That is why LinearSVC takes less time to solve the problem.

## Computational Complexity
* LinearSVC: $O(m \times n)$, based on liblinear library(no kernel trick)
* SVC: $O(m^2 \times n)$ ~ $O(m^3 \times n)$, based on libsvm library(kernel trick possible)

Therefore, SVC is suitable for small or mid size sample. And SVC is extensible to sparse features, a lot of features.

# SVM Regression
$\epsilon$ is the scale of the margin. Under the given $\epsilon$, SVR model maximizes the number of samples that fall between upper, lower margin boundaries. Thus, as long as a new sample added within the boundary, predictibility of the model does not change. (<b>$\epsilon$-insensitive$</b>) 
## Linear SVR
Use LinearSVR model

```python
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3 * X + np.random.randn(m, 1)).ravel()
```

```python
from sklearn.svm import LinearSVR
MINX, MAXX, MINY, MAXY = 0, 2, 3, 11

fig = plt.figure(figsize = (16, 6))
plot_idx = 1
epsilons = [1.5, 0.5]
X_new = np.linspace(MINX, MAXX, 100).reshape(100, 1)
for epsilon in epsilons:
    lin_svr = LinearSVR(epsilon = epsilon)
    lin_svr.fit(X, y)
    y_pred = lin_svr.predict(X_new)
    y_upper = y_pred + epsilon
    y_lower = y_pred - epsilon
    idx_off_margin = np.abs(lin_svr.predict(X) - y) >= epsilon
    plt.subplot(1, 2, plot_idx)
    plt.scatter(X[idx_off_margin], y[idx_off_margin], s = 100, alpha = 0.4, color = 'r')
    plt.scatter(X, y)
    plt.plot(X_new, y_pred, 'k-')
    plt.plot(X_new, y_upper, 'k--')
    plt.plot(X_new, y_lower, 'k--')
    plt.title('$\epsilon$ = {0}'.format(epsilon))
    plt.xlabel('X1')
    plt.ylabel('y')
    plt.axis([MINX, MAXX, MINY, MAXY])
    plot_idx += 1
```

## Non-linear SVR
Use SVR. As C is small, the model is regulated stronger.

```python
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()
```

```python
from sklearn.svm import SVR
MINX, MAXX, MINY, MAXY = -1, 1, 0, 1
EPSILON = 0.1
DEG = 2
CS = [100, 0.001]

fig = plt.figure(figsize = (16, 6))
plot_idx = 1
X_new = np.linspace(MINX, MAXX, 100).reshape(100, 1)
for C in CS:
    nonlin_svr = SVR(epsilon = EPSILON, C = C, degree = DEG, kernel = 'poly')
    nonlin_svr.fit(X, y)
    y_pred = nonlin_svr.predict(X_new)
    y_upper = y_pred + EPSILON
    y_lower = y_pred - EPSILON
    idx_off_margin = np.abs(nonlin_svr.predict(X) - y) >= EPSILON
    plt.subplot(1, 2, plot_idx)
    plt.scatter(X[idx_off_margin], y[idx_off_margin], s = 100, alpha = 0.4, color = 'r')
    plt.scatter(X, y)
    plt.plot(X_new, y_pred, 'k-')
    plt.plot(X_new, y_upper, 'k--')
    plt.plot(X_new, y_lower, 'k--')
    plt.title('degree = {0}, C = {1}, $\epsilon$ = {2}'.format(DEG, C, EPSILON))
    plt.xlabel('X1')
    plt.ylabel('y')
    plt.axis([MINX, MAXX, MINY, MAXY])
    plot_idx += 1
```
